```python
X = [75,65,80]
w = [4,3,5]
b = -1000
n = 0.01
Y = 1

```

```python
z = w[0]*X[0]+w[1]*X[1]+w[2]*X[2]+b;
```

```python
if z<0:
  y_pred=0
else:
  y_pred=1
print("Predicted output:", y_pred)

```

```python
error = Y-y_pred
if error==0:
  print("No update is required")
else:
  print("Update is required")

  for i in range(len(w)):
    w[i] = w[i]+ n*error*X[i]
```

```python
b = b+n*error
print("Updated weights:", w)
print("Updated bias:", b)
```
